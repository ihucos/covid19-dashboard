```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
df = df.groupby('Country/Region').sum().drop(columns=["Lat", "Long"]).reset_index()
```

```python
df = df.melt(id_vars='Country/Region', var_name="date")
```

```python
df.date = pd.to_datetime(df.date)
```

```python
df = df.set_index('date')
```

```python
small_deaths = df.groupby('Country/Region').sum() > 10_000
```

```python
small_deaths.head()
```

```python
merged = pd.merge(df, small_deaths, right_index=True, left_on="Country/Region")
```

```python
small_df = merged[merged.value_y == True][["Country/Region", "value_x"]]
```

```python
small_df.pivot(columns="Country/Region").plot()
```

```python
small_df.pivot(columns="Country/Region").plot()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
df = df.rename(columns={"Country/Region": "country"}).drop(["Lat", "Long"], axis=1)
```

```python
df = df.groupby("country").sum()
```

```python
df = df.reset_index().melt(id_vars="country", var_name="date").set_index("country")
```

```python
df.plot.scatter(x="date", y="value", alpha=0.1, figsize=(10, 5))
```
